**Описание полей:**

* school – школа ('GP' – Gabriel Pereira или 'MS' – Mousinho da Silveira)
* sex – пол студента ('F' – female или 'M' – male);
* age – возраст студента;
* address – домашний адрес студента ('U' – городской или 'R' – сельский);
* famsize – размер семьи ('LE3' – меньше или равно 3 или 'GT3' – больше
чем 3);
* Pstatus – статус совместного проживания родителей ('T' – совместное
проживание или 'A' – раздельное проживание);
* Medu – образование матери (0 – нет, 1 – начальное (4 класс), 2-9 классы,
3 – среднее или 4 – высшее);
* Fedu – образование отца (0 – нет, 1 – начальное (4 класс), 2-9 классы, 3 –
среднее или 4 – высшее);
* Mjob – работа матери (учитель, медицинское обслуживание, связанные с
гражданскими услугами (административными или полицейскими), на дому или
другое);
* Fjob – работа отца (учитель, медицинское обслуживание, связанные с
гражданскими услугами (административными или полицейскими), на дому или
другое);
* famrel – качество семейных отношений (от 1 – очень плохо до 5 –
отлично);
* freetime – свободное время после школы (от 1 – очень мало до 5 – очень
много);
* goout – встречи с друзьями (от 1 – очень мало до 5 – очень много);
* Dalc (Daily Alcohol Consumption) – потребление алкоголя в рабочий день
(от 1 – очень низкое до 5 – очень высокое);
* Walc (Weekly Alcohol Consumption) – еженедельное потребление
алкоголя (от 1 – очень низкое до 5 – очень высокое);
* health – текущее состояние здоровья (от 1 – очень плохое до 5 – очень
хорошее);
* absences – количество пропусков занятий в школе (от 0 до 93);
* G1 – оценка за первый период (от 0 до 20);
* G2 – оценка за второй период (от 0 до 20);
* G3 – итоговая оценка (от 0 до 20, целевой результат).

**Работа с пропущенными значениями (если они есть): удаление строк с пропущенными значениями, заполнение пропущенных значений средним значением по колонке. Если пропущенных значений нет — намеренно их "генерируете", прибивая какие-то куски данных в np.nan**

```python
stud_frame.loc[0, 'age'] = np.nan
stud_frame.loc[1, 'age'] = np.nan
stud_frame.loc[2, 'age'] = np.nan
stud_frame.loc[1, 'sex'] = np.nan
stud_frame
```

```python
stud_frame['age'] = stud_frame['age'].fillna(
    np.ceil(stud_frame['age'].mean()))
stud_frame
```

```python
stud_frame = stud_frame.dropna()
stud_frame
```

**Cоздание нового поля вычисленного на основе значений других полей:**

_через выражение на базе имеющихся колонок_

```python
stud_frame.loc[:,'Meanedu'] = (stud_frame.loc[:, 'Medu'] + stud_frame.loc[:, 'Fedu']) / 2
stud_frame
```

_через DataFrame.apply_

```python
def caculate(row):
    return (row['Medu'] + row['Fedu']) / 2

stud_frame.loc[:, 'Meanedu'] = stud_frame.apply(caculate, axis=1)
stud_frame
```

_через Series.apply_

```python
def s_calculate(value):
    return value > 12

stud_frame.loc[:, 'enrolled'] = stud_frame.loc[:, 'G3'].apply(s_calculate)
stud_frame
```

**Вычислить несколько статистик по колонкам (используйте встроенные агрегатные функции — любые на выбор)**
